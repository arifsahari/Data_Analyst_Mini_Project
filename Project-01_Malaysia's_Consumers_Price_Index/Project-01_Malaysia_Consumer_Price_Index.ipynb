##### Data Collection

Loading Dataset

The dataset is stored in a parquet file. \
It contains details such as state, date, division, and CPI index values.

```python
# Fetching dataset from official website

URL_DATA = 'https://storage.dosm.gov.my/cpi/cpi_2d_state.parquet'
df = pd.read_parquet(URL_DATA)
```

##### Data Preprocessing

- Remove null or duplicate values
- Convert data types
- Assign/Create new column based on suitable preference

```python
"""
Data Inspection
"""
# Summary
print(f'Summary Statistics: \n{df.describe()}\n')

# Dataset Shape
print(f'Total Rows: {df.shape[0]} \nTotal Columns: {df.shape[1]}\n')

# Missing Value
print(f'Missing Data per Column: \n{df.isnull().sum()}\n')

# Duplicate Value
print(f'Duplicate Data per Column: \n{df.duplicated().sum()}\n')

# Head
print(f'Head: \n{df.head()}\n')

# Info
print(f'Info: \n{df.info()}\n')
```

```python
"""
Convert data types
"""
df['index'] = df['index'].astype('int64')
df['date'] = pd.to_datetime(df['date']).dt.strftime('%Y-%m-%d')
```

```python
# Mapping 'division' codes to readable names
division_map ={
    '01': 'Food & Beverages',
    '02': 'Alcoholic Beverages & Tobacco',
    '03': 'Clothing & Footwear',
    '04': 'Housing, Utilities & Gas',
    '05': 'Household Furnishing, Equipment & Maintenance ',
    '06': 'Health',
    '07': 'Transport',
    '08': 'Information & Communication',
    '09': 'Recreation, Sport & Culture',
    '10': 'Education',
    '11': 'Restaurant & Accomodation Service',
    '12': 'Insurance & Financial Service',
    '13': 'Personal Care, Social Protection, Miscellaneous Good & Service',
    'overall' : 'Overall'
}
```

```python
"""
Assign new column
"""
# Assign new column for 'division'
df['category'] = df['division'].map(division_map)
```

##### Advanced Analysis

```python
# Import necessary library
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

Forecasting Model : \
ARIMA

```python
# Prepare data
arima_indexdate = df.set_index('date')
arima_indexdate = arima_indexdate['index'].resample('Y').mean()

# Define features and target
train_size = int(len(arima_indexdate) * 0.8)

# Split to train-test
train, test = arima_indexdate[:train_size], arima_indexdate[train_size:]
arima_model = ARIMA(train, order=(5, 1, 1)).fit()
# resample year/mean : order=511, year/sum : order=204
# resample day/sum : order=513
# resample month/mean : order=513, month/sum : order=314

# Model Forecast and Evaluation
forecast = arima_model.forecast(steps=len(test))

mae = mean_absolute_error(test, forecast)
rmse = np.sqrt(mean_squared_error(test, forecast))
r2 = r2_score(test, forecast)
print(f'MAE: {mae} \nRMSE: {rmse} \nR2: {r2}')
```

```python
# Plot results

plt.plot(train, label='Training', color='steelblue', marker='o')
plt.plot(test, label='Test', color='seagreen', marker='o')
plt.plot(test.index,forecast, label='ARIMA Forecast', color='maroon', marker='x')
plt.title(f'ARIMA Forecast vs Actual')
plt.axvline(test.index[0], label='train-test', color='black', linestyle='--')
plt.legend()
plt.grid()
plt.show()
```

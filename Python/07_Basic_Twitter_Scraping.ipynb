#### Ntscraper

```python
from ntscraper import Nitter
scraper = Nitter(log_level=1, skip_instance_check=False)
```

```python
# Scrape tweets
terms = ['github', 'JeffBezos']

tweets = scraper.get_tweets('elonmusk', mode='user', number=100)
```

```python
# Twitter profile data to scrape
profile_data = {
  'link' : [],
  'text' : [],
  'user' : [],
  'likes' : [],
  'quotes' : [],
  'retweets' : [],
  'comments' : [],
}

for tweet in tweets['tweets']:
  profile_data['link'].append(tweet['link'])
  profile_data['text'].append(tweet['text'])
  profile_data['user'].append(tweet['user']['name'])
  profile_data['likes'].append(tweet['stats']['likes'])
  profile_data['quotes'].append(tweet['stats']['quotes'])
  profile_data['retweets'].append(tweet['stats']['retweets'])
  profile_data['comments'].append(tweet['stats']['comments'])
```

```python
# View in dataframe
df = pd.DataFrame(profile_data)
df
```

```python
# Function
def scrape_twitter_data(username,number_of_tweets):
  tweets = scraper.get_tweets(username, mode='user', number=number_of_tweets)
  data = {
    'link' : [],
    'text' : [],
    'user' : [],
    'likes' : [],
    'quotes' : [],
    'retweets' : [],
    'comments' : [],
  }

  for tweet in tweets['tweets']:
    data['link'].append(tweet['link'])
    data['text'].append(tweet['text'])
    data['user'].append(tweet['user']['name'])
    data['likes'].append(tweet['stats']['likes'])
    data['quotes'].append(tweet['stats']['quotes'])
    data['retweets'].append(tweet['stats']['retweets'])
    data['comments'].append(tweet['stats']['comments'])

  df = pd.DataFrame(data)
  return df
  # df.to_csv(username+'twitter_data.csv')
```

```python
# Using the function
df_mrbeast = scrape_twitter_data('MrBeast', 1000)
df_jeffbezos = scrape_twitter_data('JeffBezos', 1000)
```

```python
# Save to csv file
import csv

with open ('scrape_csv', mode='w', newline='', encoding='utf-8') as file:
  writer = csv.writer(file)
  writer.writerow({'username', 'content'})
  for tweet in tweets['tweets']:
    writer.writerow({tweet['user']['username'], tweet['text']})
print('Tweet saved')
```

#### Snscrape

```python
import snscrape.modules.twitter as sntwitter
```

```python
attributes_container = []
```

```python
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('sex for grades since:2021-07-05 until:2022-07-06').get_items()):
    if i>150:
        break
    attributes_container.append([tweet.user.username, tweet.date, tweet.likeCount, tweet.sourceLabel, tweet.content])
```

```python
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:john').get_items()):
    if i>100:
        break
    attributes_container.append([tweet.date,
                                 tweet.likeCount,
                                 tweet.sourceLabel,
                                 tweet.content])
```

```python
df = pd.DataFrame(attributes_container, columns=["Date Created", "Number of Likes", "Source of Tweet", "Tweets"])
```

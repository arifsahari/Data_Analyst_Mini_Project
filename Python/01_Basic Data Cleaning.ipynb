```python
# Split column (cont)

# split the date and location
def split_date_location(text):
    if pd.isna(text):  # check if the value is NaN or None
        return [None, None, None]  # return empty values if it's NaN/None
    parts = re.split(r', |\(|\)', text) # split on delimiter ["," "(" ")"]
    parts = [part.strip() for part in parts if part]  # Clean up empty strings and spaces
    return parts

df[['col1', 'col2', 'col3']] = df['released'].apply(split_date_location).apply(pd.Series)
```
